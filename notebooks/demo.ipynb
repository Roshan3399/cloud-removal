# CloudClear LISS-IV Demo
Generative AI for cloud removal in LISS-IV satellite imagery

```python
import sys
sys.path.append('..')

import torch
import matplotlib.pyplot as plt
from cloudclear.models.generator import CloudRemovalGenerator
from cloudclear.data.synthetic_clouds import CloudSimulator
from cloudclear.utils.metrics import MetricsCalculator

print("All imports successful")
```

```python
# Generate synthetic test pair
sim = CloudSimulator(256)
import numpy as np
clear = np.random.rand(256, 256, 3).astype(np.float32) * 0.5 + 0.1
cloudy, mask = sim.add_clouds(clear)

fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].imshow(clear); axes[0].set_title('Clear')
axes[1].imshow(cloudy); axes[1].set_title('Cloudy')
axes[2].imshow(mask, cmap='gray'); axes[2].set_title('Cloud Mask')
plt.show()
```

```python
# Load model and run inference
model = CloudRemovalGenerator()
model.eval()

tensor = torch.from_numpy(cloudy.transpose(2, 0, 1)).unsqueeze(0).float()
with torch.no_grad():
    output, attn, density, edge = model(tensor)

output_img = output.squeeze(0).cpu().numpy().transpose(1, 2, 0)
output_img = (output_img - output_img.min()) / (output_img.max() - output_img.min())

plt.imshow(output_img)
plt.title('Cloud Removed Output')
plt.show()
```
